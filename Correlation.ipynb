Autocorrelaci√≥n

```python
import numpy as np
```

```python
roll = np.asarray(df["attitude.roll"])
roll = roll - roll.mean()
```

```python
cor = np.correlate(roll, roll, mode='full')
```

```python
plt.plot(cor)
```

```python
for i, col in enumerate(df):
    cor = np.correlate(df[col], df[col], mode='full')
    plt.subplot(6, 2, i + 1)
    plt.plot(cor)
```
