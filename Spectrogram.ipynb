```python
import os.path
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import scipy.signal
import numpy as np
```

```python
DATA_PATH = "motionsense-dataset/A_DeviceMotion_data/"
FS = 50
```

```python
df = pd.read_csv(os.path.join(DATA_PATH, "dws_1/sub_1.csv"))
del df['Unnamed: 0']
```

```python
for feat in df:
    fig = plt.figure()
    fig.suptitle(feat)
    ax1, ax2 = fig.subplots(2, 1)
    f, t, Sxx = scipy.signal.spectrogram(df[feat], fs=FS, nperseg=512, noverlap=512 * 0.95)
    ax1.pcolormesh(t, f, np.log(Sxx))
    ax2.plot(np.arange(len(df[feat])) / FS, df[feat])
```
